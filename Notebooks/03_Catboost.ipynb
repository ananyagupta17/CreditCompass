```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv('../data/loan_data_cleaned.csv')
```

```python
# Feature 1: Credit risk adjusted by loan amount
df['Credit_Loan_Ratio'] = df['Credit_History'] * df['Log_LoanAmount']

# Feature 2: Socioeconomic stability
df['Married_Educated'] = df['Married'] * df['Education']

# Feature 3: Family burden indicator
df['SelfEmp_Dependents'] = df['Self_Employed'] * df['Dependents']

# Feature 4: Burden of EMI relative to income
df['Term_Income_Ratio'] = df['Loan_Amount_Term'] / df['Log_Total_Income']

# Feature 5: Bucket loan size
# Safe binning: include lower edge and right-most value
df['LoanAmount_Bin'] = pd.cut(
    df['Log_LoanAmount'],
    bins=[0, 3, 3.5, 5],
    labels=[0, 1, 2],
    include_lowest=True,
    right=True
)

# Convert to int safely: fill NaNs first (optional strategy: fill with mode or -1)
df['LoanAmount_Bin'] = df['LoanAmount_Bin'].astype(float)  # convert to float first
df['LoanAmount_Bin'] = df['LoanAmount_Bin'].fillna(-1).astype(int)  # fill NaNs and cast to int

```

```python
# Drop target column and re-define X, y
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']

# Train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report

cat_model = CatBoostClassifier(
    iterations=300,
    learning_rate=0.03,
    depth=5,
    class_weights=[2, 1],  # Give more weight to class 0
    random_seed=42,
    verbose=0
)

cat_model.fit(X_train, y_train)

# Predict with threshold
cat_proba = cat_model.predict_proba(X_test)[:, 1]
cat_thresh = 0.36  # best threshold from your earlier tuning
cat_pred = (cat_proba >= cat_thresh).astype(int)

print("CatBoost Results:")
print(classification_report(y_test, cat_pred))
```

## SHAP — Why Does the Model Make Each Decision?

SHAP (SHapley Additive exPlanations) assigns each feature a contribution score for a given prediction. This turns the black-box model into something you can explain to a business stakeholder: "Credit History pushed the approval probability up by 0.18, but the high EMI pulled it down by 0.12."

```python
import shap
import matplotlib.pyplot as plt
import os

explainer = shap.TreeExplainer(cat_model)
shap_values = explainer.shap_values(X_test)

# For binary classification CatBoost may return a list — take class 1
sv = shap_values[1] if isinstance(shap_values, list) else shap_values

os.makedirs('../outputs', exist_ok=True)

plt.figure()
shap.summary_plot(sv, X_test, plot_type='bar', show=False)
plt.title('SHAP Feature Importance (mean |SHAP value|)')
plt.tight_layout()
plt.savefig('../outputs/shap_summary.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved to outputs/shap_summary.png')
```

```python
# Beeswarm — direction of each feature's impact per sample
# Red = high feature value, Blue = low
shap.summary_plot(sv, X_test)
```

```python
# Waterfall — explains one specific rejected loan
rejected_idx = int((y_test.values == 0).nonzero()[0][0])

base_val = explainer.expected_value
if isinstance(base_val, (list, type(None).__class__)):
    base_val = base_val[1] if isinstance(base_val, list) else base_val

shap_exp = shap.Explanation(
    values=sv[rejected_idx],
    base_values=float(base_val),
    data=X_test.iloc[rejected_idx].values,
    feature_names=X_test.columns.tolist()
)

shap.plots.waterfall(shap_exp)
```

**Key insight from SHAP:**
- `Credit_History` and `Credit_Loan_Ratio` are consistently the top drivers — a good credit history has the single biggest positive push toward approval.
- `Log_Total_Income` and `EMI` both matter significantly — higher income increases approval odds, higher EMI reduces them.
- Engineered features like `Balance_Income` and `Term_Income_Ratio` show meaningful contribution, validating our feature engineering choices.

This is directly actionable for a bank: improve credit score, reduce loan amount or increase tenure to lower EMI.

### Reading the Confusion Matrix — What Each Error Costs

For a loan model, the two types of mistakes have very different consequences:

| | Predicted: Reject | Predicted: Approve |
|---|---|---|
| **Actually: Should Reject** | ✅ True Negative | ❌ **False Positive** |
| **Actually: Should Approve** | ❌ **False Negative** | ✅ True Positive |

**False Positive (bank approves a bad loan)** → the bank disburses money to an ineligible applicant who is likely to default. Direct financial loss.

**False Negative (bank rejects a good applicant)** → the bank loses a creditworthy customer to a competitor. Lost revenue, reputational cost.

**Why we tuned threshold to 0.36 and used class_weights=[2,1]:**  
The dataset has ~70% approvals. Without correction, the model is biased toward always predicting approval — which minimises false negatives but balloons false positives. Lowering the threshold and upweighting the rejection class forces the model to be more conservative, trading a small hit to approval recall (~95%) for a meaningful gain in rejection recall (~61% → 71%). In a real bank setting, the cost of a bad loan is typically 5–10× higher than the cost of losing a good customer, so this tradeoff is deliberate and defensible.
